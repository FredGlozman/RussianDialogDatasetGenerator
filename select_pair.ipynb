```python
from lxml import etree
```

```python
xml = etree.parse('russians_rus.xml')
```

```python
root = xml.getroot()
```

```python
with open('russian_src', 'w') as fsrc, open('russian_tgt', 'w') as ftgt:
    for sess in root:
        last_utt = None
        for utt in sess:
            if last_utt is not None:
                fsrc.write(last_utt.text.strip().replace('\n', '') + '\n')
                ftgt.write(utt.text.strip().replace('\n', '') + '\n')
            last_utt = utt
```
